```python
## Import libraries ##
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import neighbors, datasets
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
```

```python
## Create a Graph to plot the data ##

iris = datasets.load_iris()

# The data we want to plote
X = iris.data[:,2:]  
y = iris.target

# Graph dimensions
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

# Width, height in inches of the graph
plt.figure(2, figsize=(8, 6))

# Clear the current figure
plt.clf()

# Plot the data points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1, edgecolor="k")

# labels
plt.xlabel(iris["feature_names"][2])
plt.ylabel(iris["feature_names"][3])
```

```python
## train_test_split- a function in Sklearn model selection for splitting data arrays into two subsets ##
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
## knn classifier ##

kclf = neighbors.KNeighborsClassifier(5)
#Classifier implementing the k-nearest neighbors vote
kclf.fit(X_train, y_train)
print(kclf.predict(X_test))
print(y_test)
kclf.score(X_test,y_test)
```

```python
## random forest classifier ##

rclf = RandomForestClassifier()
#Build a forest of trees from the training set (X, y)
rclf.fit(X_train, y_train)

print(rclf.predict(X_test))
print(y_test)
rclf.score(X_test,y_test)
```
